```python
import pandas as pd

orders_data = pd.read_csv('../data_files/retail_orders.csv', delimiter=";")

orders_data.head()
# orders_data.tail()

```

```python
orders_data[10: 15]
```

```python
orders_data.drop(
    ["ShipDate", "Status", "Country", "Note", "CustomerSegment", "MarketingTag", "InternalFlag"],
    axis=1, 
    inplace=True
)
```

```python
orders_data
```

```python
orders_data.drop(0, inplace=True)
orders_data.head()
```

```python
orders_data.reset_index(inplace=True, drop=True)
orders_data.head()
```

```python
orders_data.dtypes
```

```python
orders_data = orders_data.astype(
    {
        "Quantity":"int",
        "UnitPrice" : 'float',
        "Discount" : 'float',
        "TotalEUR": 'float'
     }
)
```

```python
orders_data.dtypes
```

```python
orders_data['OrderDatetime'] = pd.to_datetime(
    orders_data['OrderDatetime'],
    format='%Y-%m-%d %H:%M:%S',
    errors='coerce'
)

orders_data.dtypes
```

```python
orders_data.head(2)
```

```python
orders_data = orders_data[
    (orders_data["Returned"] == "False") & 
    (orders_data["OrderChannel"] == "store")
]

orders_data.head()
```

```python
orders_data = orders_data.sort_values(
    ['city', 'TotalEUR'],
    accending=[True, False]
)

orders_data
```
