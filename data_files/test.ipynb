```python
import csv
```

```python
with open ("books.csv", "r") as f:
    book_csv = csv.reader(f, delimiter=";")
    rows = list(book_csv)

    print(rows)
```

```python
titles = rows[0]
data_rows = rows[2:]
```

```python
book_collection = []

for row in data_rows:
    obj = {}
    for i in range(len(titles)):
        obj[titles[i]] = row[i]
        print(i, obj)
    book_collection.append(obj)
    book_collection[:2]
```

```python
for book in book_collection:
    if float(book["Price"]) >= 20.0 and float(book["Price"]) <= 50.0:
        print(f"Title: {book["Title"]}, Price: {book["Price"]}")
```

```python
import xml.etree.ElementTree as e
```

```python
tree = e.parse('tech_products.xml')
print(tree)

root = tree.getroot()

xml_titles = []

for i in root[0]:
    xml_titles.append(i.tag)
xml_titles
```

```python
xml_values = []

for parents in root:
    xml_value = []
    for child in parents:
        xml_value.append(child.text)
    xml_values.append(xml_value)
print(xml_values)        
```

```python
tech_collection = []

for xml_row in xml_values:
    xml_obj = {}
    for i in range(len(xml_titles)):
        xml_obj[xml_titles[i]] = xml_row[i]
    tech_collection.append(xml_obj)
tech_collection[:2]
```

```python
# REST API
import requests
```

```python
endpoint = "/api/v1/prices" 
year = "2026"
month = "02"
day = "05"
url = "https://www.sahkonhintatanaan.fi/{0}/{1}/{2}-{3}.json".format(endpoint, year, month, day)
```

```python
req = requests.get(url=url)
req.status_code
```

```python
electricity_price = req.json()
electricity_price[:2]
```

```python
highest = max(
    electricity_price,
    key=lambda i: i["EUR_per_kWh"]
)
print(f"Highest price: {highest["EUR_per_kWh"]}")
```
