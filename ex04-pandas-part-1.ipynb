1 Read the data from the csv file and create a DataFrame from it.

```python
import pandas as pd

earthquakes_data = pd.read_csv('./data_files/earthquakes.csv', delimiter=",")

earthquakes_data
```

2 Remove the following columns from the DataFrame: *Magnitude Seismic Stations*, *Azimuthal Gap*, *Horizontal Distance* and *Horizontal Error*.

```python
earthquakes_data = earthquakes_data.drop(["Magnitude Seismic Stations", "Azimuthal Gap",  "Horizontal Distance",  "Horizontal Error" ], axis=1)

earthquakes_data
```

3 Filter data from the DataFrame so that only rows matching the following criteria will be shown:

- Source must be US
- Magnitude value must be over 7.8
- ID value must start with the string "USP0007"

```python
earthquakes_data = earthquakes_data[
    (earthquakes_data["Source"] == "US") &
    (earthquakes_data["Magnitude"] > 7.8) &
    (earthquakes_data["ID"].str.startswith("USP0007"))
]

earthquakes_data
```

4 Create a new column to the <u>tenth column position</u> called Grade which will have values from the following formula: ("Magnitude" * "Depth") / "Root Mean Square". Results in this column should be presented with three decimals.

```python
grade = ((earthquakes_data["Magnitude"] * earthquakes_data["Depth"]) / earthquakes_data["Root Mean Square"]).round(3)

earthquakes_data.insert(9, "Grade", grade, True)

earthquakes_data
```

 5 Filter out events that have occurred before the 1st of January in 1996. Show the remaining rows from the DataFrame and then write the resulting DataFrame into CSV file called *filtered_earthquake_data.csv*.

```python
earthquakes_data["Date"] = pd.to_datetime(earthquakes_data["Date"])

earthquakes_data = earthquakes_data[earthquakes_data["Date"] >= "1996-01-01"]

earthquakes_data

earthquakes_data.to_csv("./data_files/filtered_earthquake_data.csv")
```
