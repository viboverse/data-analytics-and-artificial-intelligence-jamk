1 Create a function which will go through a list called **vars** and save the type of each variable to a new list. The starting information is given in the next cell.

```python
# Your function here
def get_type(lst):
    res = []
    for i in lst:
        res.append(type(i))
    return res
# Test
vars = [100,"Car",25000,10.0,["a","b","c"],True,("a","b")]

print(get_type(vars))
```

3 Below is a list containing words. Create a function which returns only words with length of 4 or less.

```python
# Your function here

def bar(list):
    result = []
    for i in list:
        if(len(i) <= 4):
            result.append(i)
    return result            


# Test
words = ["bootcamp","yellow","development","full-time","code","fresh","","static","red","advance","mill","story","are"]

print(bar(words))
```

4 Create a function which does the following:
 - Take list of dictionaries containing information of data loggers presented below as an input
 - Iterate through the dictionary and count the total amount of each sensor type. Example: {"temp":2,"hum":5,"light":2,"noise":4}
 - Return this dictionary

```python
# Your function here
def sensor_type_counts(lst):
    res = {'temp': 0, 'hum': 0, 'light': 0, 'noise': 0}

    for element in lst:
        for sensor in element['sensors']:
            if sensor in res:
                res[sensor] +=1
    return res            


# Data loggers
sdata = [{"ID":"logger-1","sensors":["temp","hum","noise"]},{"ID":"logger-2","sensors":["temp","hum"]},{"ID":"logger-3","sensors":["temp","hum","light"]},{"ID":"logger-4","sensors":["temp","hum","light"]},{"ID":"logger-5","sensors":["temp","hum","noise","light"]}]

print(sensor_type_counts(sdata))
```

5 Below is a matrix containing both letters and numbers. Create a function which does the following:
 - Collect odd numbers and then sum the collected odd numbers
 - Count the number of letters
 - Multiply the sum of odd numbers by the number of letters
 - Return the result of multiplication


```python
# Your function here

def baz(lst):
    odd = []
    letter_counts = 0

    for element in lst:
        for j in element:
            if isinstance(j, str) and j.isalpha():
                letter_counts += 1
            elif isinstance(j, int) and j % 2 != 0:
                odd.append(j)

    sum_odds = sum(odd)
    return sum_odds * letter_counts   


# Data collection
collection = [["A",10,"D",5,"H","E",3],["C",2,"F",4,7,"G",12,6],[1,3,"A","C",9],["C","E",11,8,"F",1]]

print(baz(collection))
```
